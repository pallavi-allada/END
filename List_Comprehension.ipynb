```
import random
import math
```

```
# 2.1 Add 2 iterables a and b such that a is even and b is odd
a = random.sample(range(1,100),k=15)
b = random.sample(range(1,100),k=15)
sun_func = lambda a,b:[x+y for x,y in zip(a,b) if x%2==0 and y%2!=0]
print(a)
print(b)
sun_func(a,b)
```

```
# 2.2 Strips every vowel from a string provided (tsai>>t s)
vowel = ['a','e','i','o','u']
strip_func = lambda strng:" ".join([ch for ch in strng if not ch in vowel])
strip_func("tsai")
```

```
# 2.3 Acts like a ReLU function for a 1D array
relu_func = lambda arr:[max(0, x) for x in arr]
relu_func([0,12,-1,7])
```

```
# 2.4 Acts like a sigmoid function for a 1D array
sigmoid_func = lambda arr:[1/(1 + math.exp(-x)) for x in arr]
sigmoid_func([2,7,0.6,0,434567])
```

```
# 2.5 Takes a small character string and shifts all characters by 5 (handle boundary conditions) tsai>>yxfn
alphabets = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
shift_func= lambda strng:"".join([alphabets[alphabets.index(x) + 5] if alphabets.index(x) < 21 else alphabets[alphabets.index(x) - len(alphabets) + 5] for x in strng])    
shift_func('tsai'), shift_func('yak')
```
