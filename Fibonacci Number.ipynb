```
#Function to generate a list with Fibonacci numbers

def generateFib(n:int):
  n1=0
  n2=1
  try:
    if isinstance(n,int) and n>0:
      fib = [n1,n2]
      while n1+n2 < n:
        fib.append(n1+n2)
        n1,n2=n2,n1+n2
      return fib
    else:
      raise Exception("Not a valid integer")
  except Exception as e:
      print(e)

fibs = generateFib(10000)
```

```
#Function to check if a number is fibonacci or not
def isFib(n:int, fibs:list):
  if len(list(filter(lambda x: x in fibs, [n]))) > 0:
    return "Fibonacci"
  return "Not Fibonacci"
```

```
user_number = input("Enter number to check if Fibonacci or not")
print(user_number,"-->",isFib(user_number,fibs))
```
