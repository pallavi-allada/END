```
from functools import reduce
```

```
# 4. Using reduce function:
# 4.1 Add only even numbers in a list
sum_func = lambda lst:reduce(lambda x, y: x + y if y%2 ==0 else x, lst,0)
sum_func([1,2,3,4,5,6])
```

```
# 4.2 find the biggest character in a string (printable ascii characters)
biggest_char_func = lambda strng: reduce(lambda x,y: x if ord(x) > ord(y) else y, strng)
biggest_char_func("shg768$%")
```

```
# 4.3 Adds every 3rd number in a list
sum_3rd_func = lambda lst: reduce(lambda x,y:x+y, lst[2::3])
sum_3rd_func([1,2,3,4,5,6,6,7,8,9,8])
```
