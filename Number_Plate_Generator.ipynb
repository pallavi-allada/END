```
# 5.Using randint, random.choice and list comprehensions, write an expression that generates 15 random KADDAADDDD number plates, 
# where KA is fixed, D stands for a digit, and A stands for Capital alphabets. 10<<DD<<99 & 1000<<DDDD<<9999 PTS:100
```

```
import random
```

```
state_code = 'KA'
KA_number_plates = [state_code+str(random.randint(10,99))+chr(random.randint(65,90))+chr(random.randint(65,90))+str(random.randint(1000,9999)) for i in range(0,15)]
KA_number_plates
```

```
# 6.Write the above again from scratch where KA can be changed to DL, and 1000/9999 ranges can be provided.  PTS:100
def num_plate_generator(state_code = 'KA',start=1000,end=9999):
  return [state_code+str(random.randint(10,99))+chr(random.randint(65,90))+chr(random.randint(65,90))+str(random.randint(1000,9999)) for i in range(0,15)]
num_plate_generator('DL'), num_plate_generator('KA')
```
